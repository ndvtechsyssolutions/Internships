```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('matches.csv')
```

```python
print(df.isnull().sum())
```

```python
#drop the method column
df = df.drop(columns=['method'])
#drop the  super_over column
df = df.drop(columns=['super_over'])
df
```

```python
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns   
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='match_type', order=df['match_type'].value_counts().index)
plt.title('Count of Matches by Type')
plt.xlabel('Match Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='winner', order=df['winner'].value_counts().index)
plt.title('Count of Matches Won by Each Team')
plt.xlabel('Count')
plt.ylabel('Winning Team')
plt.show() 
```

```python
plt.figure(figsize=(10, 8))
sns.countplot(data=df, y='city', order=df['city'].value_counts().index)
plt.title('Count of Matches by City')
plt.xlabel('Count')
plt.ylabel('City')
plt.show()
```

```python
#count of played matches by year
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='year', order=df['year'].value_counts().index)
plt.title('Count of Matches Played by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
df.to_csv('cleaned_matches.csv', index=False)
```
