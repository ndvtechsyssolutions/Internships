```python
#load dataset
import pandas as pd
import numpy as np

df=pd.read_csv('/content/drive/MyDrive/Data/netflix_titles.csv')
df
```

```python
print("INFO:")
df.info()

print("\nDESCRIBE:")
print(df.describe())

print("\nMISSING VALUES:")
print(df.isnull().sum())

print("\nDUPLICATES:")
print(df.duplicated().sum())
```

```python
#handling missing values
df['director'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df.dropna(subset=['title'], inplace=True)
```

```python
#remove duplicates
df.drop_duplicates(inplace=True)
```

```python
#convert datatypes
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
```

```python
#using numpy
import numpy as np
df['is_old'] = np.where(df['release_year'] < 2010, True, False)
```

```python
#use pandas to filter
# Filter: only TV Shows
tv_shows = df[df['type'] == 'TV Show']

# Sort by release year
df.sort_values(by='release_year', ascending=False).head()

# Group by country
df.groupby('country')['title'].count().sort_values(ascending=False).head()
```

```python
#summary statistics
print("Summary of Numeric Columns:")
print(df.describe())

print("\nSummary of Object/Text Columns:")
print(df.describe(include='object'))

print("\nCount of Content Types (Movie vs TV Show):")
print(df['type'].value_counts())

print("\n Count of Content Ratings:")
print(df['rating'].value_counts())

print("\nTop 10 Countries by Number of Titles:")
print(df['country'].value_counts().head(10))
```
