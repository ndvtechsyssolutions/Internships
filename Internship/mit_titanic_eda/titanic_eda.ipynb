### Visualizations 

Survival Count:

0 - Did Not Survive

1 - Survived

Insight:

Less than half of the passengers survived

```python
sns.countplot(x='Survived', hue='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

Embarked Count:

0, 1, 2 - the location from which the passengers boarded the titanic

Insight:

Most of the passengers embarked from location no. 2.0
While location 1.0 had least amount of embarked passengers 

```python
sns.countplot(x='Embarked', data=df)
plt.title("Embarked Count")
plt.show()
```

Survival Count by the Embarkment Point:

0.0, 1.0, 2.0 - location of embarkment

Legend(upper right):

0 - Did not Survive

1 - Survived

Insight:

Embarked Location 2.0 had the most casualties

while other embarked location also had low survival count

```python
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title("Survival Count by Embarkment Point")
plt.show()
```

Count of Passengers by Distribution of Age

Insight:

The passenger with age near 28 had the highest count

The passengers with age more than 60 were quite low

```python
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

Survival Status of Passengers by Age

Insight:

Every age distribution had impact on survival

While the most impacted one was around age 28 which had highest passenger count by age

```python
# Create the histplot of Age, differentiating by Survived status
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Survived', kde=True, multiple='stack', palette='viridis')
plt.title("Age Distribution by Survival Status")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend(title='Survived', labels=['Survive', 'Did Not Survived'])
plt.show()
```

Survival Count of Passengers by Gender

Gender:

0 - Male

1 - Female

Color:

Blue - Did not survive

Orange - Survived

Insight:

Survival Count of males was quite low compared to females

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

A boxplot showing the survival by age

Color:

Blue - Did not survive

Orange - Survived

Insight:

It follows similar trend as that of age survival histogram

While also showing some few outliers in both Did not survived and Survived

```python
sns.boxplot(x='Survived', y='Age', hue='Survived', data=df)
plt.title("Age vs Survival")
plt.show()
```

A Correlation Heatmap of the titanic dataset

Insight:

Gender had relatively high positive correlation with survival

Number of siblings or spouses aboard had relatively high positive correlation with number of parents or children aboard.

Passenger Class and Fare had relatively high negative correlation

```python
plt.figure(figsize=(8,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
