```python
#import all Required library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.dropna(subset=['cast'], inplace=True) #drop the row ,where cast value missing
df['date_added'] = df['date_added'].str.strip() #clean space and convert to datetime 
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df=df.dropna(subset=['date_added'])   #Drop the row date_added with invalid or misising value 
df['director'].fillna('Unknown',inplace=True)  #fill the misisng value in director column with Unknowm
df['country'].fillna('Unknown',inplace=True)  # fill the missing value with unknown in country column
df['rating']=df['rating'].fillna(df['rating'].mode()[0])  #fill the rating missing value with mode value 
df['duration']=df['duration'].fillna(df['duration'].mode()[0]) #fill the duration misiing value with the mode value
```

```python
df.isnull().sum() #Again Checking is any  Misisng value present 
```

```python
type_counts = df['type'].value_counts()  #Count the Value by Content Type 
print("\nContent Type Counts:")
print(type_counts)
```

```python
plt.figure(figsize=(12, 6))  #plotting the Missing Value Graph
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Value Heatmap")
plt.show()
```

```python
sns.countplot(data=df, x='type', palette='Set2') #plotting the Content type Distribution Graph
plt.title("Content Type Distribution")
plt.show()
```

```python
df.to_csv("clean_netflix_titles.csv", index=False) # Expoting the clean csv File
print("\nCleaned dataset saved as 'clean_netflix_titles.csv'")
```
